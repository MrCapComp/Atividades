```
n1 = int(input("digite algum nnumero pra saber se é multiplo de 2: "))
n2 = int(input("digite algum nnumero pra saber se é multiplo de 3: "))
if n1 % 2 == 0 and n2 % 3 == 0:
  print(f"o numero {n1} é multiplo e 2, e o {n2} é multiplo de 3")
elif n1 % 2 == 0 and n2 % 3 == 1:
  print(f"o numero {n1} é multiplo por 2, mas o numero {n2} não é multiplo de 3")
elif n1 % 2 == 1 and n2 % 3 == 0:
  print(f"o numero{n1} não é multiplo de 2, mas o {n2} é multiplo de 3")
elif n1 % 2 == 1 and n2 % 3 == 1:
  print(f"o numero {n1} nem o {n2} são multiplos")
else:
  print("insira um numero por favor")
```

```
n1 = int(input("digite o primeiro nnumero da adição"))
n2 = int(input("digite p segundo numero da adição"))

while n1 != 0:
  n3 = n1 + n2
  print(f"sua soma deu {n3}")
  n1 = int(input("digite o primeiro nnumero da adição"))
  n2 = int(input("digite p segundo numero da adição"))

print("operação finalizada")
```
