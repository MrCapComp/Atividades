```
#-- Menu
print("--- Sistema de Fechamento de Caixa ---\n\nBem vindo ao sistema simples de fechamento de caixa, \nonde iremos pedir-lhe a quantidade de notas e iremos somar tudo \nautomaticamente para você.\n")





dinheiro_caixa = {
  "100 reais": 100,
  "50 reais": 50,
  "20 reais": 20,
  "10 reais": 10,
  "5 reais": 5,
  "2 reais": 2,
  "1 real": 1,
  "0,50 cents": 0.50,
  "0,25 cents": 0.25,
  "0,10 cents": 0.10,
  "0,05 cents": 0.05
}
total_caixa = 0

resultados={}
total_geral = 0

for nome, valor in dinheiro_caixa.items():
  qtd = int(input(f"Quantas notas de {nome} vc tem? "))
  subtotal = qtd * valor
  print(f"Subtotal de {nome}: RS{subtotal}")
  total_caixa += subtotal

resultados[nome] = subtotal
total_geral += subtotal
print(f"\n####-- Fechamento de caixa de hoje: R${total_geral:.2f} --####\n")




```
